```python
%%writefile main1.cpp
#include <bits/stdc++.h>
using namespace std;


int solveQuartic(double a, double b, double c, double x[]) {
    if (a == 0 && b == 0 && c == 0) {
        return -1;
    }

    if (a == 0 && b == 0) {
        return 0;
    }

    if (a == 0) {
        double y = -c / b;
        if (y < 0) return 0;
        x[0] = sqrt(y);
        x[1] = -sqrt(y);
        return 2;
    }

    double delta = b * b - 4 * a * c;
    if (delta < 0) return 0;

    double y1 = (-b + sqrt(delta)) / (2 * a);
    double y2 = (-b - sqrt(delta)) / (2 * a);

    int count = 0;
    if (y1 >= 0) {
        x[count++] = sqrt(y1);
        x[count++] = -sqrt(y1);
    }
    if (y2 >= 0 && y2 != y1) {
        x[count++] = sqrt(y2);
        x[count++] = -sqrt(y2);
    }

    return count;
}

// ---------------- Test framework ----------------

struct TestCase {
    double a, b, c;
    int expected_n;
    vector<double> expected_roots;
    string description;
};

bool compare_roots(vector<double> got, vector<double> expected, double tol=1e-6) {
    if (got.size() != expected.size()) return false;
    sort(got.begin(), got.end());
    sort(expected.begin(), expected.end());
    for (size_t i=0; i<got.size(); i++) {
        if (fabs(got[i] - expected[i]) > tol) return false;
    }
    return true;
}

int main() {
    vector<TestCase> tests = {
        {0,0,0, -1, {}, "Trường hợp kiểm thử 1: Vô số nghiệm"},
        {0,0,1, 0, {}, "Trường hợp kiểm thử 2: Vô nghiệm (a=0,b=0,c!=0)"},
        {0,2,-8, 2, {2.0,-2.0}, "Trường hợp kiểm thử 3: a=0,b!=0,y>0"},
        {0,2,0, 2, {0.0,0.0}, "Trường hợp kiểm thử 4: a=0,b!=0,y=0"},
        {0,2,8, 0, {}, "Trường hợp kiểm thử 5: a=0,b!=0,y<0"},
        {1,0,-1, 2, {1.0,-1.0}, "Trường hợp kiểm thử 6: a!=0,delta>0,1 y>0,1 y<0"},
        {1,0,1, 0, {}, "Trường hợp kiểm thử 7: a!=0,delta<0"},
        {1,-2,1, 2, {1.0,-1.0}, "Trường hợp kiểm thử 8: a!=0,delta=0,y>0"},
        {1,1,0, 2, {0.0,0.0}, "Trường hợp kiểm thử 9: a!=0,1 y=0"},
        {1,-5,4, 4, {2.0,-2.0,1.0,-1.0}, "Trường hợp kiểm thử 10: a!=0,delta>0,2 y>0"},
        {2,-3,1, 4, {1.0,-1.0,sqrt(0.5),-sqrt(0.5)}, "Trường hợp kiểm thử 11: a!=0,delta>0,2 y>0, sqrt không nguyên"}
    };

    int pass_count = 0;
    for (size_t i=0; i<tests.size(); i++) {
        auto t = tests[i];
        double x[10];
        int n = solveQuartic(t.a,t.b,t.c,x);
        vector<double> got_roots;
        if (n > 0) {
            for (int j=0;j<n;j++) got_roots.push_back(x[j]);
        }

        bool ok = (n == t.expected_n);
        if (ok && n > 0 && t.expected_roots.size() > 0) {
            ok = compare_roots(got_roots, t.expected_roots);
        }

        cout << t.description << "\n";
        cout << "Đầu vào: a="<<t.a<<", b="<<t.b<<", c="<<t.c<<"\n";
        cout << "Kết quả mong đợi n="<<t.expected_n;
        if (!t.expected_roots.empty()) {
            cout << ", nghiệm=";
            for (auto v : t.expected_roots) cout << v << " ";
        }
        cout << "\nKết quả nhận được n="<<n;
        if (n > 0) {
            cout << ", nghiệm=";
            for (auto v : got_roots) cout << v << " ";
        }
        cout << "\nKết quả: " << (ok ? "PASS":"FAIL") << "\n\n";
        if (ok) pass_count++;
    }

    cout << "Tóm tắt: " << pass_count << " / " << tests.size() << " trường hợp kiểm thử đã qua.\n";
    return 0;
}
```
